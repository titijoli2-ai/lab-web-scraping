```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Check how many books are displayed

books = soup.find_all("article", class_="product_pod")

len(books)
```

```python
# Request the book page
book_response = requests.get(full_url)

# Check that everything worked
book_response.status_code
```

```python
# Parse the HTML of the book detail page
book_soup = BeautifulSoup(book_response.content, "html.parser")
```

```python
# Extract the Universal Product Code (UPC)
upc = book_soup.find("table").find_all("td")[0].text

upc
```

```python
# Extract the book genre from the breadcrumb navigation
genre = book_soup.find_all("a")[3].text

genre
```

```python
# Extract the product description
description = book_soup.find("meta", attrs={"name": "description"})["content"].strip()

description
```

```python
# Create an empty list to store book information
books_data = []

# Loop through all books on the page
for book in books:

    # Extract the title
    title = book.h3.a["title"]

    # Extract and clean the price
    price = float(book.find("p", class_="price_color").text.replace("£", ""))

    # Extract and convert the rating
    rating_text = book.find("p", class_="star-rating")["class"][1]
    rating = rating_map[rating_text]

    # Extract availability
    availability = book.find("p", class_="instock availability").text.strip()

    # Build the detail page URL
    book_url = book.h3.a["href"]
    full_url = "https://books.toscrape.com/" + book_url

    # Request and parse the detail page
    detail_response = requests.get(full_url)
    detail_soup = BeautifulSoup(detail_response.content, "html.parser")

    # Extract UPC
    upc = detail_soup.find("table").find_all("td")[0].text

    # Extract genre
    genre = detail_soup.find_all("a")[3].text

    # Extract description
    description = detail_soup.find("meta", attrs={"name": "description"})["content"].strip()

    # Store all information
    books_data.append({
        "UPC": upc,
        "Title": title,
        "Price (£)": price,
        "Rating": rating,
        "Genre": genre,
        "Availability": availability,
        "Description": description
    })
```

```python
def scrape_books(min_rating, max_price):
    # Website page to scrape
    url = "https://books.toscrape.com/"
    
    # Send request and parse HTML
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Find all book containers
    books = soup.find_all("article", class_="product_pod")
    
    # Dictionary to convert text ratings into numbers
    rating_map = {
        "One": 1,
        "Two": 2,
        "Three": 3,
        "Four": 4,
        "Five": 5
    }
    
    # Store scraped data
    books_data = []
    
    # Loop through all books on the page
    for book in books:
        title = book.h3.a["title"]
        price = float(book.find("p", class_="price_color").text.replace("£", ""))
        
        rating_text = book.find("p", class_="star-rating")["class"][1]
        rating = rating_map[rating_text]
        
        availability = book.find("p", class_="instock availability").text.strip()
        
        book_url = book.h3.a["href"]
        full_url = "https://books.toscrape.com/" + book_url
        
        detail_response = requests.get(full_url)
        detail_soup = BeautifulSoup(detail_response.content, "html.parser")
        
        upc = detail_soup.find("table").find_all("td")[0].text
        genre = detail_soup.find_all("a")[3].text
        description = detail_soup.find("meta", attrs={"name": "description"})["content"].strip()
        
        # Apply filters
        if rating >= min_rating and price <= max_price:
            books_data.append({
                "UPC": upc,
                "Title": title,
                "Price (£)": price,
                "Rating": rating,
                "Genre": genre,
                "Availability": availability,
                "Description": description
            })
    
    # Return results as a DataFrame
    return pd.DataFrame(books_data)
```

```python
filtered_books = scrape_books(min_rating=4, max_price=20)

filtered_books
```
